# Saving and manipulating thebeat's plots

---

## Saving plots
In the example below we create a recurrence plot, and save it to disk using ``fig.savefig()``. We ask for a plot with a dpi of 600, so we get better quality.

```python
import warnings
warnings.filterwarnings('ignore', message="Matplotlib ")
%matplotlib inline
```

```python
from thebeat import Sequence
from thebeat.visualization import recurrence_plot

seq = Sequence(iois=[300, 400, 300, 400])
fig, ax = recurrence_plot(seq, dpi=600)
fig.savefig('recurrence_plot.png')
```

## Manipulating plots

```python
fig, ax = recurrence_plot(seq, suppress_display=True)
ax.set_xlabel('Event number')
ax.set_ylabel('Event number')
fig.show()
```
