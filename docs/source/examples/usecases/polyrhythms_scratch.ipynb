# Polyrhythms and syncopation
In this notebook we can do stuff with polyrhythms (possibly the incredibly complicated thing from peter vuust etc.) and/or the keller conference paper with the simple notes.

 PRODUCTION AND SYNCHRONIZATION OF UNEVEN RHYTHMS AT FAST TEMPI this one

```python
from thebeat.rhythm import Rhythm
from thebeat.core import Sequence, SoundStimulus, SoundSequence
```

```python
# We suppress warnings, but let's hide that to avoid confusion
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
rhythm = Rhythm.from_integer_ratios([2, 2, 3], time_signature=(7, 8))
rhythm.plot_rhythm(staff_type="rhythm", dpi=600);
```

---

```python
rhythm = Rhythm.from_integer_ratios([2, 3, 2, 3], time_signature=(5, 8))
rhythm.plot_rhythm(staff_type="rhythm", print_staff=False, dpi=600);
```

```python
r = Rhythm.generate_random_rhythm()
```

---

We can now also do syncopation stuff with ties between bars:

```python
rhythm = Rhythm.from_integer_ratios([1, 2, 2, 2, 1])
rhythm.plot_rhythm(staff_type="rhythm", dpi=600);
```

---

```python
rhythm = Rhythm.from_integer_ratios([1, 2, 1])
rhythm.plot_rhythm(staff_type="rhythm", dpi=600);
```

---

```python
rhythm = Rhythm.from_note_values([16, 16, 16, 4, 2, 4, 4, 4, 4, 16])
rhythm.plot_rhythm('syncopated_rhythm.png', dpi=600)
```

---

If we want to listen to such highly syncopated rhythms, it might be useful to have a metronome sound:

```python
seq = rhythm.to_sequence()
stim = SoundStimulus.generate()
trial = SoundSequence(stim, seq)
trial.play(metronome=True)
```

```python
# To get the soundplayer to work on this website we need the following. Please ignore it. On your computer just execute the block above.
import thebeat.helpers
from IPython.display import Audio
samples = thebeat.helpers.get_sound_with_metronome(trial.samples, trial.fs, trial.mean_ioi, 1.0)
Audio(samples, rate=trial.fs)
```
