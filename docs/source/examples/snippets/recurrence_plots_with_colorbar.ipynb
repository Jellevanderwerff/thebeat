# Plot two recurrence plots with colorbar

```python
from thebeat.core import Sequence, Stimulus
import matplotlib.pyplot as plt
import matplotlib as mpl
from thebeat.visualization import recurrence_plot

fig, axs = plt.subplots(1, 2)

seq1 = Sequence.generate_random_normal(n=10, mu=500, sigma=50, metrical=True) * 10
seq2 = Sequence.generate_random_normal(n=10, mu=500, sigma=50, metrical=True) * 10

recurrence_plot(seq1, ax=axs[0])
recurrence_plot(seq1, ax=axs[1])

fig.colorbar(mpl.cm.ScalarMappable(), ax=axs, fraction=0.02, pad=0.1)
fig.show()
```
