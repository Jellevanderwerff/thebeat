# Combining and repeating objects
*thebeat* contains a number of functions for manipulating existing objects. These are useful when combining different types of sequences, or when you want to elongate an existing object.

```python
from thebeat.core import Sequence, Stimulus, StimSequence
from thebeat.rhythm import Rhythm
```

```python
# We suppress warnings, but let's hide that to avoid confusion
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Initial sequence
seq = Sequence.generate_random_normal(n=10, 
                                      mu=500, 
                                      sigma=50, 
                                      metrical=True)
seq.plot_sequence();

# Repeat sequence 10 times and plot
seq_longrepeats = seq * 10
fig, ax = seq_longrepeats.plot_sequence();
```

---

```python
rhythm1 = Rhythm.from_integer_ratios([1, 1, 2])
rhythm2 = Rhythm.from_integer_ratios([2, 1, 1])
rhythm = rhythm1 + rhythm2
rhythm.plot_rhythm(dpi=600, staff_type="percussion");
```
